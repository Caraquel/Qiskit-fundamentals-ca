## Library Imports

## Single-Qubit Circuit Example
This cell builds and displays a simple 1-qubit quantum circuit with a Hadamard gate and measurement.

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1)

qc.h(0)

qc.measure_all()

qc.draw(output='mpl')
```

## Two-Qubit Circuit Test
This cell creates a 2-qubit circuit, applies an H gate to the first qubit, entangles the two qubits with a CNOT, and measures both qubits.

```python
from qiskit import QuantumCircuit

qc2 = QuantumCircuit(2)
qc2.h(0)
qc2.cx(0, 1)
qc2.measure_all()

qc2.draw(output='mpl')
```
